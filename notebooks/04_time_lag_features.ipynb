# Time-Lag Feature Engineering for Regime Prediction

Decide to create two prediction variables, short-term: $y_{t+5}$, and medium/long-term: $y_{t+20}$. 

In our dataset, we have the following mapping of segments to regimes:
- **Segment 0** := BEAR
- **Segment 1** := BULL

```python
# Load the data
import pandas as pd
FILE_PATH = r'C:\Users\jacob\Documents\GitHub\COMP0047-Data-Science-Project\notebooks\data\labeled_dataset.csv'
df = pd.read_csv(FILE_PATH)
df.drop(columns=['segment'], inplace=True)
df.head()
```

```python
# Map regime to binary values
regime_mapping = {'bull': 1, 'bear': 0}
df['target'] = df['regime'].map(regime_mapping)
df['target'].value_counts()
```

```python
h_long = 20
h_short = 5

# Shift the prediction variables
df['target_long'] = df['target'].shift(-h_long)
df['target_short'] = df['target'].shift(-h_short)
df.head()
```

```python
# Drop rows with NaN values in the target columns
df.dropna(subset=['target_long', 'target_short'], inplace=True)
df.isna().sum()
```
